**Importing libraries:**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importing CSV file and identify the columns name**

```python
data = pd.read_csv("/content/haberman.csv", names=['age', 'year_of_treatment', 'positive_lymph_nodes', 'survival_5yr'])
```

```python
data.shape # checking the shape of the data
```

```python
data.info() # checking the identified columns names
```

```python
data.head() # checking data
```

```python
data.isnull().sum() # checking missing values or null values
```

```python
data.describe() # Describe my data
```

```python
data['survival_5yr'].value_counts()
```

**Considering in survival that yes == 1 and no == 2**

```python
data['survival_5yr'].replace(to_replace = 1, value ="yes", inplace = True)
data['survival_5yr'].replace(to_replace = 2, value ="no", inplace = True)
data['survival_5yr'].value_counts()
```

#Lets start with visulaization

**Univariate analysis**

```python
sns.countplot(x = data['survival_5yr']).set_title('Count Plot for Survival_5yr')
plt.show()
```

```python
plt.hist(data.age,color = 'blue',bins = 20)
plt.show()
```

```python
sns.boxplot(data = data,x='age')
plt.show()
```

```python
plt.figure(figsize=[12,8])
plt.subplot(2,2,1)
sns.boxplot(data = data,y='year_of_treatment')
plt.subplot(2,2,(3,6))
sns.boxplot(data = data,x='positive_lymph_nodes')
plt.show()
```

**Multi-variate analysis**

```python
plt.figure(figsize=[20,15])
sns.pairplot(data,hue = "survival_5yr")
plt.show()
```

**Bivariate analysis**

```python
sns.jointplot(x="age", y="year_of_treatment", data=data, color ='green')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x="age", y="year_of_treatment", data=data, hue = "survival_5yr")
plt.show()
```

```python
sns.jointplot(x="age", y="year_of_treatment", data=data, hue = "survival_5yr")
plt.show()
```

```python
sns.catplot(x= 'survival_5yr', y = 'age', data = data, kind = "box")
plt.show()
```

```python
sns.catplot(x= 'survival_5yr', y = 'year_of_treatment', data = data, kind = "boxen")
plt.show()
```

```python
sns.violinplot( x = 'survival_5yr', y = 'age', data = data)
plt.show()
```

##Final conclusions :-


###* Patients age and operation year alone are not deciding factors for his/her survival. Yet, people less than 35 years have more chance of survival.
###* Survival chance is inversely proportional to the number of positive axillary nodes. We also saw that the absence of positive axillary nodes cannot always guarantee survival.
###* The objective of classifying the survival status of a new patient based on the given features is a difficult task as the data is imbalanced
